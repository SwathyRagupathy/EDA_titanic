```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("markmedhat/titanic")

print("Path to dataset files:", path)
```

```python
csv_file_path = path + "/titanic.csv"
```

```python
# Import necessary libraries
import pandas as pd  # For data manipulation
import numpy as np  # For numerical computations
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For advanced visualizations
```

```python
data = pd.read_csv(csv_file_path)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Basic dataset information
print("\n--- Dataset Information ---")
data.info()  # Overview of data types and missing values
```

```python
# Summary statistics for numerical features
print("\n--- Summary Statistics ---")
print(data.describe())  # Summary of numerical columns
```

```python
# Checking for missing values
print("\n--- Missing Values ---")
print(data.isnull().sum())  # Count of missing values for each column
```

```python
# Visualize missing values using a heatmap
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Check for duplicate rows
duplicates = data.duplicated().sum()
print(f"\n--- Duplicate Rows: {duplicates} ---")
```

```python
# Distribution of target variable (Survived)
sns.countplot(data=data, x="Survived", palette="pastel")
plt.title("Distribution of Survival")
plt.show()
```

```python
# Univariate Analysis
## Distribution of Age
sns.histplot(data['Age'].dropna(), kde=True, bins=30, color='blue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
## Passenger Class Distribution
sns.countplot(data=data, x="Pclass", palette="viridis")
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# Bivariate Analysis
## Survival by Gender
sns.countplot(data=data, x="Sex", hue="Survived", palette="Set2")
plt.title("Survival by Gender")
plt.show()
```

```python
## Survival by Passenger Class
sns.countplot(data=data, x="Pclass", hue="Survived", palette="Set3")
plt.title("Survival by Passenger Class")
plt.show()
```

```python
## Age vs. Fare Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived", palette="coolwarm")
plt.title("Age vs. Fare (Colored by Survival)")
plt.show()
```

```python
# Handling Missing Values
## Filling missing Age values with the median
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
## Filling missing Embarked values with the mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
# Encoding Categorical Variables for Visualization
## Mapping genders to numeric values
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
```

```python
# Outlier Detection
## Boxplot for Fare
sns.boxplot(data=data, x="Fare")
plt.title("Boxplot for Fare")
plt.show()
```

```python
# Pairplot to visualize relationships between numerical features

sns.pairplot(data[['Age', 'Fare', 'Pclass', 'Survived']], hue="Survived", palette="husl")
plt.show()
```

```python
# Feature Engineering
## Creating a Family Size feature
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
sns.countplot(data=data, x="FamilySize", hue="Survived", palette="coolwarm")
plt.title("Survival by Family Size")
plt.show()
```

```python
## Creating an Age Group feature
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 12, 20, 40, 60, 80], labels=["Child", "Teen", "Adult", "Middle Age", "Senior"])
sns.countplot(data=data, x="AgeGroup", hue="Survived", palette="viridis")
plt.title("Survival by Age Group")
plt.show()
```

```python
# Save the cleaned dataset for future use
data.to_csv("titanic_cleaned.csv", index=False)
print("\n--- Cleaned dataset saved as titanic_cleaned.csv ---")
```
